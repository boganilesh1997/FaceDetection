```python
# This is cascades to defect the face obtained by trainging our model.
face = cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/haarcascade_frontalface_default.xml')
```

```python
# This detect method used to detect face and return the image with ploting squre around the detected face 
def detect(gray,frame):
  faces = face.detectMultiScale(gray,1.3,5)
  for (x,y,w,h) in faces:
    cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),5)
    onlyFace = frame[y:y+h, x:x+w]
  return frame,onlyFace
```

```python
img = io.imread("/content/drive/MyDrive/Colab Notebooks/image.jpg")
plt.imshow(img)
```

Image which will be provided as imput to the method 

```python
#converting image to Gray Scale image.
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

Image after converting into Gray Scale

```python
"""Now we will use the method which is created to detect face,
This method will accept the 2 input images 
1) Actula image -> img 
2) Gray Scale Image -> gray
and Return the image with face detected 
"""
Face_detected,onlyface = detect(gray,img)
plt.imshow(Face_detected)
plt.imshow(onlyface)
```
